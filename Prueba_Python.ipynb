```python
# Importamos las librerías a usar:
import control
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creamos la función transferencia:
num = [24]
den = [1 , 17 , 32 , 30]
sys = control.TransferFunction(num,den)
```

```python
# Calculamos polos y ceros
control.bode_plot(sys)
print(control.pole(sys))
print(control.zero(sys))
control.pzmap(sys)
```

```python
# Calculamos la respuesta al escalón:
t, y = control.step_response(sys)
plt.figure(3)
plt.plot(t, y, 'r')
plt.ylabel('Xp(t)')
plt.xlabel('Xin(t)')
```

```python
# Finalmente obtenemos las matrices de estado:
print(control.tf2ss(sys))
```
