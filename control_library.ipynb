 En el presente archivo se probrán las funciones que vienen en la librería "control" para python. Para ello utilizamos el ejercicio 11 de la guía 2. El objetivo es generar una función transferencia, graficar su respuesta al escalón, diagrama de bode y polos/ceros.

```python
#Importamos las librerías a usar

import control
import matplotlib.pyplot as plt
import numpy as np

# Creamos vectores que contienen los coeficientes del numerador y denominador

num = [0.7 , 2]
den = [1 , 1.5 , 2]

# Generamos la función transferencia

sys = control.TransferFunction(num,den)
print(sys)
```

```python
# Para mostrar las figuras en una nueva ventana o en la consola, escribir en la terminal del spyder:
# Escribir en la consola!         %matplotlib auto
# Escribir en la consola!         %matplotlib inline    para escribir en la consola


# Calculamos polos y ceros

print(control.pole(sys))
print(control.zero(sys))
```

```python
# Ploteamos los polos y ceros

control.pzmap(sys)
```

```python
# Graficamos el bode

control.bode_plot(sys)
```

```python
# Calculamos respuesta al escalón

t, y = control.step_response(sys)
plt.figure(3)
plt.plot(t, y, 'r')
plt.ylabel('Xp(t)')
plt.xlabel('Xin(t)')
```

```python
# A partir de la ft, obtenemos las matrices A,B,C,D

print(control.tf2ss(sys))
```
